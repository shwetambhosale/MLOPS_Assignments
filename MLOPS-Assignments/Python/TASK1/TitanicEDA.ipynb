```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# TitanicEDA class
class TitanicEDA:
    def __init__(self, dataset_path):
        self.dataset_path = dataset_path
        self.df = None
    
    # Method to load the data
    def load_data(self):
        self.df = pd.read_csv(self.dataset_path)
        print(f"Data loaded successfully from {self.dataset_path}")
    
    # Method to generate summary statistics
    def summary_statistics(self):
        print("Summary Statistics:")
        print(self.df.describe())
        print("\nMissing Data:")
        print(self.df.isnull().sum())
    
    # Method to visualize the distribution of survival rates by 'Pclass'
    def plot_survival_by_pclass(self):
        plt.figure(figsize=(6,4))
        sns.countplot(data=self.df, x='Pclass', hue='Survived')
        plt.title('Survival Distribution by Pclass')
        plt.savefig('survival_by_pclass.png')
        plt.show()
    
    # Method to visualize the distribution of survival rates by 'Sex'
    def plot_survival_by_sex(self):
        plt.figure(figsize=(6,4))
        sns.countplot(data=self.df, x='Sex', hue='Survived')
        plt.title('Survival Distribution by Sex')
        plt.savefig('survival_by_sex.png')
        plt.show()

    # Method to visualize the distribution of survival rates by 'Age'
    def plot_survival_by_age(self):
        plt.figure(figsize=(6,4))
        sns.histplot(data=self.df, x='Age', hue='Survived', multiple='stack', bins=20)
        plt.title('Survival Distribution by Age')
        plt.savefig('survival_by_age.png')
        plt.show()
    
    # Method to visualize the correlation matrix
    def plot_correlation_matrix(self):
        plt.figure(figsize=(8,6))
        correlation = self.df.corr()
        sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
        plt.title('Correlation Matrix')
        plt.savefig('correlation_matrix.png')
        plt.show()
```

```python
# Create an instance of TitanicEDA class
titanic_eda = TitanicEDA('titanic.csv')
```

```python
# Load the dataset
titanic_eda.load_data()
```

```python
# Generate summary statistics
titanic_eda.summary_statistics()
```

```python
# Visualize the distribution of survival rates by 'Pclass'
titanic_eda.plot_survival_by_pclass()
```

```python
# Visualize the distribution of survival rates by 'Sex'
titanic_eda.plot_survival_by_sex()
```

```python
# Visualize the distribution of survival rates by 'Age'
titanic_eda.plot_survival_by_age()
```

```python
# Updated method to visualize the correlation matrix with numeric columns only
def plot_correlation_matrix(self):
    # Select only numeric columns
    numeric_df = self.df.select_dtypes(include=['number'])
    
    # Compute the correlation matrix
    plt.figure(figsize=(8,6))
    correlation = numeric_df.corr()
    
    # Plot the correlation matrix
    sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix')
    plt.savefig('correlation_matrix.png')
    plt.show()
```
